# Do it for the Chi Squared

```python
# Calculate sample means for 100 distributions of variabe sample sizes
distributions['chi2'] = [
    np.random.chisquare(df=mymean, size=N).mean()
    for N in sample_sizes
]

# Scatter Plot as a function of sample size
chisqrplt = plt.figure().add_subplot(111)
chisqrplt.plot(sample_sizes, distributions['chi2'],'o')
chisqrplt.plot([min(sample_sizes), max(sample_sizes)], [100, 100], 'k')
plt.title('Chi-Squared Distribution Mean as related to Sample Size')
plt.xlabel('Sample Size')
plt.ylabel('Sample Mean');
```

Figure 1: The sample mean for 100 Chisq distributions against the increasing sample size N (independent var). The sample mean  gets nearer to the population mean (100) or parent distribution as the sample size of the distributions inceases (Law of large numbers - Central Limit Theorem).


```python
plt.hist(distributions['chi2'],bins=30)
plt.title('Sample means distribution of 100 Chi squared Distributions')
plt.xlabel('sample mean', fontsize = 18)
plt.ylabel('N', fontsize = 18)
```

Figure 2: The histrogram showing the distribution of Chi squared sample means. The figure shows that the larger the sample size, the closer sample mean is to the population mean.
