```python
# One Date/Time and one numberical column is picked up to plot the scattor plot
x_column, y_column = 'DATE', 'AMNT'

# 5. Remove all the numberical columns except the 'DATE' & 'AMNT' column
df_datetime = df.drop([
        column for column in df.columns 
        if column not in [x_column, y_column]
], axis=1).dropna(how='any')


df_datetime[x_column] = pd.to_datetime(df_datetime[x_column])

#Reduced dataframe
df_datetime
```

```python
df_datetime.set_index('DATE').groupby(pd.TimeGrouper('D')).sum().dropna()
```
